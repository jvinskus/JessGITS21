```python
cities=['Seattle', 'San Francisco', 'Los Angeles', 'Denver', 'Salt Lake City', 'Phoenix', 'Chicago', 'St. Louis', 'Dallas', 'Boston', 'Washington D.C.', 'Miami']
tday=[306,314,275,366,366,366,366,366,306,306,366,275]
pday=[14,22,119,34,21,70,28,10,24,0,3,2]
aqimed=[42,45,81,58,50,75,56,52,49,42,42,44]
zone= ['west','west', 'west', 'mountain','mountain','mountain', 'central',  'central', 'central', 'east', 'east', 'east']
```

```python
import pandas as pd
```

```python
aqidict={'City':cities, 'Total Days':tday, 'Poor Days':pday, 'AQI Median':aqimed, 'Zone':zone}
```

```python
aqicities=pd.DataFrame(aqidict)
aqicities
```

```python
aqicities2=pd.read_csv('aqicities2.csv')
aqicities2
```

```python
allcities=pd.concat([aqicities,aqicities2], axis=0, join='inner')
allcities
```

```python
allcities=allcities.reset_index()
allcities= allcities.drop(['index'], axis=1)
allcities
```
