```python
import geopandas as gp
import pandas as pd
```

```python
path = gp.datasets.get_path('nybb')
```

```python
gdf= gp.read_file(path)
```

```python
gdf #coordnates in state plane coordinates, 
#shows all the pts that make up the polygon. 
```

```python
gdf.crs #coordinate referencing system
```

```python
gdf.plot()
```

```python
gdf['area']=gdf.area
gdf
```

```python
gdf['centroid']=gdf.centroid
gdf
```

```python
firstpoint=gdf['centroid'].iloc[0]
gdf['distance']= gdf['centroid'].distance(firstpoint)
gdf
```

```python
gdf['distancemiles']= gdf['distance']/5280
gdf
```

```python
cities= pd.DataFrame({'City':['Paris', 'Tokyo','Buenos Ares','Los Angeles'], 
                      'Lon': [2.352, 139.650, -58.381, -118.241],
                      'Lat': [48.856, 35.676, -34.605, 34.057]})
cities
```

```python
citygdf=gp.GeoDataFrame(cities, geometry=gp.points_from_xy(cities['Lon'], cities['Lat']))
citygdf
```
