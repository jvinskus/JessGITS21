```python
import pandas as pd
```

```python
cities=['Seattle', 'San Francisco', 'Los Angeles', 'Denver', 'Salt Lake City', 'Phoenix', 'Chicago', 'St. Louis', 'Dallas', 'Boston', 'Washington D.C.', 'Miami']
tday=[306,314,275,366,366,366,366,366,306,306,366,275]
pday=[14,22,119,34,21,70,28,10,24,0,3,2]
aqimed=[42,45,81,58,50,75,56,52,49,42,42,44]
zone= ['west','west', 'west', 'mountain','mountain','mountain', 'central',  'central', 'central', 'east', 'east', 'east']
aqidict20={'City':cities, 'Total Days':tday, 'Poor Days':pday, 'AQI Median':aqimed, 'Zone':zone}
aqicities20=pd.DataFrame(aqidict20)
aqicities20
```

```python
cit20 =['Pittsburg', 'Syracuse']
tot20 = [366,366]
por20 = [14,0]
aqimed20= [49,36]
zone20=['east','east']
aqidict220={'City':cit20, 'Total Days':tot20, 'Poor Days':por20, 'AQI Median':aqimed20, 'Zone':zone20}
aqicities220=pd.DataFrame(aqidict220)
aqicities220
```

```python
cities10=['Seattle', 'San Francisco', 'Los Angeles', 'Denver', 'Salt Lake City','Phoenix', 'Chicago', 'St. Louis', 'Dallas', 'Boston', 'Washington D.C.', 'Miami']
tday10=[365,365,365,365,365,  365,365,365,365,365,365,365]
pday10=[3,14,115,38,23,       57,43,   103,38,10,50,6]
aqimed2010=[41,48,77,61,51,   74,71,   74, 52,76,55,48]
aqidict10={'City':cities10, 'Total Days 2010':tday10, 'Poor Days 2010':pday10, 'AQI Median 2010':aqimed2010}

aqicities10=pd.DataFrame(aqidict10)
aqicities10
```

```python
cit10 =['Pittsburg', 'Syracuse']
tot10 = [365,365]
por10 = [83,6]
aqimed10= [69,37]
zone10=['east','east']
aqidict210={'City':cit20, 'Total Days 2010':tot10, 'Poor Days 2010':por10, 'AQI Median 2010':aqimed10}
aqicities2010=pd.DataFrame(aqidict210)
aqicities2010
```

```python
#all 2020
allcities20=pd.concat([aqicities20,aqicities220], axis=0, join='inner')
allcities20
allcities20=allcities20.reset_index()
allcities20= allcities20.drop(['index'], axis=1)
allcities20
```

```python
#all 2010
allcities10=pd.concat([aqicities10,aqicities2010], axis=0, join='inner')
allcities10=allcities10.reset_index()
allcities10= allcities10.drop(['index'], axis=1)
allcities10
```

```python
alldata= pd.merge(allcities20,allcities10,how='right')
alldata
```

```python
alldata['Scale 2020']=alldata['Total Days']/366
#used 366 bc leap year
```

```python
alldata['Poor Day Rate 2020']=alldata['Poor Days']/alldata['Scale 2020']
alldata['Poor Day Rate 2020']=alldata['Poor Day Rate 2020'].astype(int)
alldata
```

```python
alldata['Scale 2010']=alldata['Total Days 2010']/365
alldata['Poor Day Rate 2010']=alldata['Poor Days 2010']/alldata['Scale 2010']
alldata['Poor Day Rate 2010']=alldata['Poor Day Rate 2010'].astype(int)
alldata
```

```python
import numpy as np
citr=np.arange(len(alldata['City']))
```

```python
import matplotlib.pyplot as plt
# part A
plt.style.use('fivethirtyeight')
plt.bar(citr-0.15,alldata['Poor Day Rate 2010'],color='blue', label='2010', width=0.3)
plt.bar(citr+0.15,alldata['Poor Day Rate 2020'],color='red', label='2020', width=0.3)
plt.xlabel('City')
plt.ylabel('Poor Days')
plt.grid(True)
plt.xticks(ticks = citr, labels = alldata['City'], rotation='vertical' )
plt.legend()
plt.title('Poor AQI Rate by City')
```

```python
#part B
plt.style.use('fivethirtyeight')
plt.bar(citr-.15,alldata['AQI Median 2010'],color='blue', label='2010', width=0.3)
plt.bar(citr+0.15,alldata['AQI Median'],color='red', label='2020', width=0.3)
plt.xlabel('City')
plt.ylabel('AQI')
plt.grid(True)
plt.xticks(ticks = citr, labels = alldata['City'], rotation='vertical' )
plt.legend()
plt.title('AQI Median by City')
```
