# example of a new geodataframe

```python
df = pd.DataFrame(
    {'City': ['Buenos Aires', 'Brasilia', 'Santiago', 'Bogota', 'Caracas'],
     'Country': ['Argentina', 'Brazil', 'Chile', 'Colombia', 'Venezuela'],
     'Latitude': [-34.58, -15.78, -33.45, 4.60, 10.48],
     'Longitude': [-58.66, -47.91, -70.66, -74.08, -66.86]})
df
```

```python
gdf = gp.GeoDataFrame(df, geometry=gp.points_from_xy(df.Longitude, df.Latitude))
gdf
```

```python
world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))

# We restrict to South America.
ax = world[world.continent == 'South America'].plot(color='white', edgecolor='black')

# We can now plot our ``GeoDataFrame``.
gdf.plot(ax=ax, color='red')

plt.show()
```

```python
gdf.to_file("maps/SA_cities.shp")
```

```python
path = gp.datasets.get_path('nybb')
df = gp.read_file(path)
df.head()
```

```python
df.plot(figsize=(6, 6))
plt.show()
```

```python
print(df.crs)
```

```python
df = df.to_crs(epsg=4326)
print(df.crs)
df.head()
```

```python
m = folium.Map(location=[40.70, -73.94], zoom_start=10, tiles='CartoDB positron')
m
```

```python
for _, r in df.iterrows():
    #without simplifying the representation of each borough, the map might not be displayed
    #sim_geo = gpd.GeoSeries(r['geometry'])
    sim_geo = gp.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['BoroName']).add_to(geo_j)
    geo_j.add_to(m)
m
```

```python
aqistats= pd.DataFrame({'City':city, 'Total Days':totaldays, 'Poor Days':poordays, 'AQI Median':aqimed, 'Zone':zone, 'Lon':lon, 'Lat':lat})
```

```python
gdf=gp.GeoDataFrame(aqistats, geometry=gp.points_from_xy(aqistats.Lon,aqistats.Lat))
gdf
```
